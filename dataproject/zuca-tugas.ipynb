```python
import pandas as pd
import numpy as np
```

```python
#Importing data from
url = "https://api.statbank.dk/v1/data/INDKP104/CSV?lang=en&LANDDEL=*&ENHED=116&SOCIO=*&INDKOMSTTYPE=100%2C105%2C110&Tid=%3E2006"
```

```python
data = pd.read_csv(url, delimiter=";")
```

```python
df= data[data["LANDDEL"] == "All Denmark"] #Selecting the region "All Denmark"
```

```python
#organizing data and table
pd.pivot_table(df, values= 'INDHOLD', index=['TID', "ENHED", "SOCIO", "LANDDEL"], columns=['INDKOMSTTYPE']).reset_index()
```

```python
#selecting Students in All Denmark
dfs=df[(df['SOCIO']=='Students')] 
dfs
```

```python
#reorganize the data into the different stages of income
dfs1 = pd.pivot_table(dfs, values= 'INDHOLD', index=['TID', "ENHED", "SOCIO", "LANDDEL"], columns=['INDKOMSTTYPE']).reset_index()
dfs1
```
